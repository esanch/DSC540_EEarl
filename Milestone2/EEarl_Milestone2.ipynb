```python
import pandas as pd
df = pd.read_csv('IMDB-Movie-Data.csv')
df.head()
#Clean data: columns to be dropped
#Description & Runtime (unnecessary for determining its popularity)
```

```python
df.drop(columns=['Description', 'Runtime (Minutes)'], inplace=True)
df.head()
```

```python
#clutter exist in actors and as no spacing between commas in some rows
for i, v in enumerate(df['Genre'].values):
    v = v.replace(',', ', ') #replace all commas with a comma and space after
    v = v.replace(', ', ', ')
    df['Genre'].values[i] = v
df.head()
```

```python
for i, v in enumerate(df['Actors'].values):
    v = v.replace(',', ', ') #replace all commas with a comma and space after
    v = v.replace(', ', ', ')
    df['Actors'].values[i] = v
df.head()
```

```python
#In order to have a clean merge with other tables later I want to eliminate "The" if it is at the beginning of any movie title
#typically when in the beginning The is capitalized so we rely on exact case matching
for i, v in enumerate(df['Title'].values):
    v = v.replace('The ', '')
    v = v.replace('The', '')
    df['Title'].values[i] = v
df.head(10) #row 5 (rank 6) has 'The Great Wall' which we want changed to Great Wall. Showing top 10 rows will check if change is correctly made
```

```python
#Although the number of votes help determine the film's popularity it is an arbitrary number as how many total votes were casted. 
    #For this reason we drop this column that gives unneeded info
df.drop(columns=['Votes'], inplace=True)
df.head()
```

```python
#Title isnt as explict so Title will be replaced with "Film title"
df.rename(columns={'Title': 'Film Title'}, inplace=True)
df.head()
```

```python
#Save changes back to file
df.to_csv('IMDB_MovieData_Copy.csv')
```

```python
from bs4 import BeautifulSoup as bs
import pandas as pd 
with open("websiteData.html", "r", encoding="utf8") as fd:
    soup = bs(fd)
    print("There are {} tables".format(len(soup.find_all("table"))))
print(soup.find_all("table"))
```

```python
#THE FOLLOWING CODE WAS TAKEN FROM https://www.geeksforgeeks.org/convert-html-table-into-csv-file-in-python/
#TO MAKE HTML FILE INTO CSV FILE FOR (EASIER) FURTHER CLEANING
import os
import sys
import pandas as pd
from bs4 import BeautifulSoup
   
path = 'websiteData.html'
   
# empty list
data = []
   
# for getting the header from
# the HTML file
list_header = []
soup = bs(open(path, encoding="utf8"),'html.parser')
header = soup.find_all("table")[0].find("tr")
  
for items in header:
    try:
        list_header.append(items.get_text())
    except:
        continue
  
# for getting the data 
HTML_data = soup.find_all("table")[0].find_all("tr")[1:]
  
for element in HTML_data:
    sub_data = []
    for sub_element in element:
        try:
            sub_data.append(sub_element.get_text())
        except:
            continue
    data.append(sub_data)
  
# Storing the data into Pandas
# DataFrame 
dataFrame = pd.DataFrame(data = data, columns = list_header)
   
# Converting Pandas DataFrame
# into CSV file
dataFrame.to_csv('cleanWebsiteData.csv')
```

```python
df1 = pd.read_csv('cleanWebsiteData.csv')
df1.head()
```

```python
#Distibutor and "Unnamed: 0" are unnecessary so columns can be dropped
df1.drop(columns=['Unnamed: 0', 'Distributor'], inplace=True)
df1.head()
```

```python
#Similar to previous CSV data I will remove any leading The in Movie column and rename "Movie" column to "Film Title"
for i, v in enumerate(df1['Movie'].values):
    v = v.replace('The ', '')
    v = v.replace('The', '')
    df1['Movie'].values[i] = v
df1.rename(columns={'Movie': 'Film Title'}, inplace=True)
df1.head()
```

```python
#My CSV file has '2021 Gross' as 'Revenue (Millions)' and 'ReleaseDate' as 'Year' so I will change the column names 
df1.rename(columns={'2021 Gross': 'Revenue (Millions)'}, inplace=True)
df1.rename(columns={'ReleaseDate': 'Year'}, inplace=True)
```

```python
#I have to ensure Revenue value is correct format so $ character must be dropped
df1['Revenue (Millions)'] = [str(x).replace('$', '') for x in df1['Revenue (Millions)']]
df1.head()
```

```python
#Year must be in correct format currently it has month and day which I do not need
df1['Year'] = df1['Year'].astype('datetime64[Y]')
```

```python
df1['Year']= pd.DatetimeIndex(df1['Year']).year
df1['Year'] = [str(x).replace('.0', '') for x in df1['Year']]
df1.head()
#year column is corrent now with only the year
```

```python
df1.to_csv('cleanWebsiteDataCopy.csv')
```

```python
#THE FOLLOWING CODE WAS TAKEN FROM https://www.geeksforgeeks.org/convert-json-to-csv-in-python/
#TO MAKE HTML FILE INTO CSV FILE FOR (EASIER) FURTHER CLEANING
import json
import csv
 
# Opening JSON file and loading the data
# into the variable data
with open('apiData.json') as json_file:
    data = json.load(json_file)
 
employee_data = data['items']
 
# now we will open a file for writing
data_file = open('cleanApiDataCopy.csv', 'w')
 
# create the csv writer object
csv_writer = csv.writer(data_file)
 
# Counter variable used for writing
# headers to the CSV file
count = 0
 
for emp in employee_data:
    if count == 0:
 
        # Writing headers of CSV file
        header = emp.keys()
        csv_writer.writerow(header)
        count += 1
 
    # Writing data of CSV file
    csv_writer.writerow(emp.values())
 
data_file.close()
```

```python
df2 = pd.read_csv('cleanApiDataCopy.csv')
df2.head()
```

```python
#unnecessary columns are dropped
df2.drop(columns=['id', 'fullTitle', 'image', 'imDbRatingCount'], inplace=True)
```

```python
#Once again removing leading 'The' in titles
for i, v in enumerate(df2['title'].values):
    v = v.replace('The', '')
    v = v.replace('The ', '')
    df2['title'].values[i] = v
```

```python
#Fixing headers to match previous dataTables
df2.rename(columns={'title': 'Film Title'}, inplace=True)
df2.rename(columns={'rank': 'Rank'}, inplace=True)
df2.rename(columns={'year': 'Year'}, inplace=True)
df2.rename(columns={'imDbRating': 'Rating'}, inplace=True)
df2.head()
```

```python
#creating new columns to split the 'crew' column into directors/actors
df2['Director']=''
df2['Actors'] =''
df2.head()
```

```python
#splitting on the first comma as director is the first name
new = df2['crew'].str.split(",", n = 1, expand = True)
df2["Director"]= new[0]
df2["Actors"]= new[1]
df2['Director'] = df2['Director'].str.split("(",expand=True) #this will drop the (dir.), portion in the column
df2.head()
```

```python
#once the new columns are made I can drop the crew column as it is not needed
df2.drop(columns=['crew'], inplace=True)
df2.head()
```

```python
#clean actors column with wrongfully placed commas
for i, v in enumerate(df2['Actors'].values):
    v = v.replace(',', ', ')
    v = v.replace(', ', ', ') #replace all commas with a comma and space after
    df2['Actors'].values[i] = v
df2.head()
```
